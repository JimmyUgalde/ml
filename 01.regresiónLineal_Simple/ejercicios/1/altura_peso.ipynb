```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#https://gist.github.com/nstokoe/7d4717e96c21b8ad04ec91f361b000cb

datos = pd.read_csv("weight-height.csv")
#prueba con otro dataset
#datos = pd.read_csv("table.csv")
datos.info()
datos.head()
```

```python
sns.scatterplot(x="Height", y="Weight", data=datos, hue="Weight",palette="coolwarm")
plt.show()
```

```python
X = datos["Height"]
Y = datos["Weight"]

X_procesada = X.values.reshape(-1,1)
Y_procesada = Y.values.reshape(-1,1)
```

```python
modelo = LinearRegression()

modelo.fit(X_procesada,Y_procesada)
```

```python
centimetros = 178

pulgadas = centimetros / 2.54

#pulgadas = 73.847017

prediccion = modelo.predict([[pulgadas]])

print("libras: ",round(prediccion[0][0]))

kilos = prediccion[0][0] * 0.45359237

print("kilos: ",round(kilos))

#Prueba
#pulgadas = 73.847017
#resultado = 241 / 219 = 22lib (10.05%)
```

```python
coeficiente = modelo.score(X_procesada,Y_procesada)

por_coeficiente = coeficiente * 100

print("Coeficiente de determinación :", round(por_coeficiente, 2), "%")
```
